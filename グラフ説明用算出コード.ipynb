顧客単価の初月から最後の月までの変化率・差算出

```python
diff_c_unit =  month_all_df['顧客単価'].iloc[-1] - month_all_df['顧客単価'].iloc[0]
pct_c_unit = (diff_c_unit / month_all_df['顧客単価'].iloc[0])*100
```

総売上高の初月から最後の月までの変化率・差算出

```python
diff_sales =  month_all_df['総売上高'].iloc[-1] - month_all_df['総売上高'].iloc[0]
pct_sales = (diff_sales / month_all_df['総売上高'].iloc[0])*100
```

顧客数の初月から最後の月までの変化率・差算出

```python
diff_c =  month_all_df['顧客合計'].iloc[-1] - month_all_df['顧客合計'].iloc[0]
pct_c = (diff_c / month_all_df['顧客合計'].iloc[0])*100
```

```python
weekly_all_df['リピート率']
```

リピート率の土日と平日の平均値の差

```python
weekly_all_df['リピート率'].iloc[5:]
```

```python
weekly_all_df['リピート率'].iloc[0:5]
```

```python
diff_r = weekly_all_df['リピート率'].iloc[0:5].mean() - weekly_all_df['リピート率'].iloc[5:].mean()
```

顧客数の土日と平日の平均値の差

```python
diff_c_w =  weekly_all_df['顧客合計'].iloc[5:].mean() - weekly_all_df['顧客合計'].iloc[0:5].mean()
```

出力

```python
print('顧客単価の変化率は',np.round(pct_c_unit,3),'%')
print('顧客単価の差は',diff_c_unit,'円')
print('総売上高の変化率は',np.round(pct_sales,3),'%')
print('総売上高の差は',diff_sales,'円')
print('顧客数の変化率は',np.round(pct_c,3),'%')
print('顧客数の差は',diff_c,'人')
print('土日と平日のリピート率平均値の差',diff_r,'%')
print('土日と平日の顧客数平均値の差',diff_c_w,'人')
```
